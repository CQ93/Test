#### edge 생성

```python
G = nx.Graph() # 비어있는 구조 생성.
```

```python
G.add_edge('IU','EPIK HIGH') # edge 생성
nx.draw(G, with_labels = True, font_weigth = "bold") # label 보여주기, 폰트 굵게.
plt.show()
```

```python
# IU와 LOCO edge 추가
G.add_edge('IU','LOCO')
nx.draw(G, with_labels = True, font_weigth = "bold")
```

```python
G.remove_edge('IU','EPIK HIGH')
nx.draw(G, with_labels = True, font_weigth = "bold")
#plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 12))

# 1. 그래프 생성
g = nx.from_pandas_edgelist(df, source='to', target='from') 

# 2. 레이아웃
layout = nx.spring_layout(g,iterations=50)

# 3. 세부적인 조정
From_size = [g.degree(F) * 80 for F in From]
nx.draw_networkx_nodes(g, 
                       layout, 
                       nodelist=From, 
                       node_size=From_size, # a LIST of sizes, based on g.degree
                       node_color='lightblue')

# 
nx.draw_networkx_nodes(g, layout, nodelist=To, node_color='blue', node_size=100) # 

# 노드 많은거 주황색으로 표시
popular_To = [To for To in To if g.degree(To) > 1]
nx.draw_networkx_nodes(g, layout, nodelist=popular_To, node_color='orange', node_size=100) # 연결점 많은 것들 주황색으로

nx.draw_networkx_edges(g, layout, width=1, edge_color="gray")

node_labels = dict(zip(From, From))
nx.draw_networkx_labels(g, layout, labels=node_labels)

#겉에 선 테두리 없애기.
plt.axis('off')

#제목
plt.title("Tweet Follower ...")

# 끝
plt.show()
```
